# Creating Training Data and Test Data

```python
with open('train.txt', 'r') as file:
    data = file.read().splitlines()
```

```python
data
```

```python
len(data)
```

```python
label = []
for path in data:
    disease = path.split('/')[1]
    if disease == 'Functionaldysphonia':
        label.append(0)
    elif disease == 'Neoplasm':
        label.append(1)
    elif disease == 'Phonotrauma':
        label.append(2)
    else:
        label.append(3)
```

```python
train_data = pd.DataFrame({'filepath': data, 'label': label})
train_data
```

```python
with open('dev.txt', 'r') as file:
    dev_data = file.read().splitlines()
```

```python
label = []
for path in dev_data:
    disease = path.split('/')[1]
    if disease == 'Functionaldysphonia':
        label.append(0)
    elif disease == 'Neoplasm':
        label.append(1)
    elif disease == 'Phonotrauma':
        label.append(2)
    else:
        label.append(3)
```

```python
eval_data = pd.DataFrame({'filepath': dev_data, 'label': label})
eval_data
```

```python
train_data.to_csv('csv_files/train_data_not_processed.csv')
```

```python
eval_data.to_csv('csv_files/eval_data_not_processed.csv')
```
